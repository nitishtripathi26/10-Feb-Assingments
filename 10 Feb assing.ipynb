Question 1-Which function is used to open a file? What are the different modes of opening a file? Explain each mode
of file opening.

```python
Ans-In Python, the open() function is used to open a file.

The open() function takes two arguments - the file name (or file path) and the mode in which the file should be opened.
The mode parameter is optional and if not specified, the file is opened in read-only mode by default.

Here are the different modes of opening a file:

1-'r' - Read mode: This is the default mode of opening a file. The file is opened in read-only mode, and the file pointer
is placed at the beginning of the file. If the file does not exist, an error is raised.

2-'w' - Write mode: In this mode, the file is opened in write mode, and the file pointer is placed at the beginning 
of the file. If the file already exists, its contents are truncated. If the file does not exist, a new file is created.

3-'a' - Append mode: In this mode, the file is opened in append mode, and the file pointer is placed at the end of the file.
If the file already exists, new data is appended to the end of the file. If the file does not exist, a new file is created.

4-'x' - Exclusive mode: In this mode, the file is opened in exclusive mode, and a new file is created. If the file already 
exists, an error is raised.

5-'b' - Binary mode: In this mode, the file is opened in binary mode, which is used for reading and writing binary data
such as images or audio files.

6-'t' - Text mode: In this mode, the file is opened in text mode, which is used for reading and writing text data.

Example of Opening a file in a write mode:
    
# open a file in write mode
file = open("example.txt", "w")

# write some data to the file
file.write("This is a sample text.\n")

# close the file
file.close()

In this example, we use the 'open()' function to open a file named example.txt in write mode by passing the mode parameter
as "w". We then write some text data to the file using the write() method, and finally close the file using the close()
method.
```

Question 2-Why close() function is used? Why is it important to close a file?

```python
Ans-In Python, the close() function is used to close a file. Its important to close a file after reading from or
writing to it to free up system resources and prevent any data loss or corruption.

Here are some reasons why its important to close a file:

Free up system resources: When a file is opened, the operating system allocates system resources to the file.
If the file is not closed properly, these system resources are not released, and the program may run out of resources,
causing the system to slow down or crash.

Save data: When you write to a file, the data is first stored in a buffer in memory. If the file is not closed, the data
in the buffer may not be written to the file, resulting in data loss or corruption.

Prevent security issues: If a file is not closed, it may be possible for other programs or users to access it, leading
to potential security issues.

# Open the file
file = open("example.txt", "r")

# Read the contents of the file
data = file.read()

# Close the file
file.close()

By closing the file, we ensure that any resources associated with the file are released, and the data is properly 
written to the file if it was being written.
```

Question 4-Explain the following with python code: read(), readline() and readlines().

```python
Ans-In Python, there are several methods available to read the contents of a file. Here's an explanation of the read(), 
readline(), and readlines() methods, along with examples of how to use them:

1-'read()': This method reads the entire contents of a file and returns it as a string.
Example:
    # Open the file in read mode
file = open("data.txt", "r")

# Read the entire contents of the file
contents = file.read()

# Close the file
file.close()

# Print the contents of the file
print(contents)

2-'readline()': This method reads one line at a time from a file and returns it as a string. When called again, it
reads the next line.
Example:
    # Open the file in read mode
file = open("data.txt", "r")

# Read the first line of the file
line1 = file.readline()

# Read the second line of the file
line2 = file.readline()

# Close the file
file.close()

# Print the lines of the file
print(line1)
print(line2)

3-'readlines()': This method reads all the lines of a file and returns them as a list of strings.
Example:
    # Open the file in read mode
file = open("data.txt", "r")

# Read all the lines of the file
lines = file.readlines()

# Close the file
file.close()

# Print the lines of the file
for line in lines:
    print(line)

In the above examples, we first open the file in read mode using the open() function. Then, we use the appropriate method 
('read()', 'readline()', or 'readlines()') to read the contents of the file. Finally, we close the file using
the close() method.

```

Question 5-Explain why with statement is used with open(). What is the advantage of using with statement and
open() together?

```python
In Python, the with statement is used to open a file and automatically close it when the block of code is exited.
The open() function is used to open a file for reading or writing.

with open("data.txt", "r") as file:
    contents = file.read()
    print(contents)

The with statement is used to open the file named "data.txt" in read mode using the open() function.
The file object is assigned to the variable file. Within the with block, we read the contents of the file using the 
read() method and print it to the console. When the block of code is exited, the file is automatically closed.

The advantage of using the with statement with the open() function is that it ensures that the file is properly closed,
even if an error occurs while the file is being read or written. This is because the with statement guarantees that the
file will be closed as soon as the block of code is exited, regardless of how the block is exited.
```

Question 6-Explain the write() and writelines() functions. Give a suitable example.

```python
Ans-In Python, the write() and writelines() functions are used to write to a file. Here's an explanation of each function,
along with examples of how to use them:

1-'write()': This function is used to write a single string to a file. If the file does not exist, it will be created.
Example:
    
# Open the file in write mode
file = open("data.txt", "w")

# Write a string to the file
file.write("I want to become a data scientist")

# Close the file
file.close()

2-'writelines()': This function is used to write a list of strings to a file. If the file does not exist, it will
be created.
Example:

# Open the file in write mode
file = open("data.txt", "w")

# Write a list of strings to the file
lines = ["I want to become a data scientist", "I love Python"]
file.writelines(lines)

# Close the file
file.close()

In Above examples, we first open the file in write mode using the open() function. Then, we use the appropriate function
(write() or writelines()) to write to the file. Finally, we close the file using the close() method.
```
